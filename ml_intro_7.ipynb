### Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
study_hours = [3,4,5,6,7,8,9,2,1]
print(study_hours)
print(len(study_hours))
score = [60,75,80,85,80,88,90,45,30]
print(score)
print(len(score))
```

```python
stu = pd.DataFrame({
    "hours" : study_hours,
    "score" : score
})
stu
```

```python
sns.scatterplot(stu,x=stu["hours"],y=stu["score"])
```

```python
corr = stu.corr(method="pearson")
```

```python
corr
```

```python
as we know in pearson corelation coefficent we always get value between -1 to 1, as value close to 1 or -1 we can say they are
more corelated, as value close to 0 we can say they are not more corelated.

from the above we can say study-hours is more corelated to the score in the makrs as we can see pearson corelation coff value :
    0.914564 which is close to 1.
    
so we can say hours is positively corelated wiht score, if the value of hours is increace the value of score will be increase
if we decrease the value of hours then score will also be decrease.
```

### Q3. Suppose you are conducting a study to examine the relationship between the number of hours of exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation between these two variables and compare the results.

```python
hours = np.random.randint(1,11,50)
body_mass = np.random.randint(1,5,50)
print(hours)
print(body_mass)
```

```python
df = pd.DataFrame({"hours":hours,"body_mass_index":body_mass})
df
```

```python
corr_1 = df.corr(method="pearson")
corr_2 = df.corr(method="spearman")
print(corr_1)
print(corr_2)
```

```python
pear = corr_1.iloc[1,0]
spear = corr_2.iloc[1,0]
print(pear,spear)
```
